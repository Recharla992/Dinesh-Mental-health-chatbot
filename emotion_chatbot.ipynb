```python
import pickle

# Load trained model
with open("sentiment_model.pkl", "rb") as f:
    sentiment_model = pickle.load(f)

with open("tfidf_vectorizer.pkl", "rb") as f:
    vectorizer = pickle.load(f)

print("‚úÖ Model and Vectorizer Loaded")
```

```python
def predict_sentiment(text):
    vec = vectorizer.transform([text])
    pred = sentiment_model.predict(vec)[0]
    prob = sentiment_model.predict_proba(vec).max() * 100

    prob = float(prob)   # üî• FIX: convert numpy ‚Üí python float

    if prob < 60:
        label = "Neutral"
    elif pred == 1:
        label = "Positive"
    else:
        label = "Negative"

    return label, round(prob, 2)
```

```python
predict_sentiment("I feel very low today")
```

```python
from transformers import pipeline

emotion_ai = pipeline(
    "text-classification",
    model="j-hartmann/emotion-english-distilroberta-base",
    top_k=3,
    framework="pt"
)

def detect_emotion(text):
    emotions = emotion_ai(text)[0]
    return [(e["label"], round(e["score"]*100, 2)) for e in emotions]


# Test
print(detect_emotion("I feel scared and lonely"))
```

```python
EMERGENCY_WORDS = [
    "suicide", "kill myself", "want to die",
    "no reason to live", "end my life"
]

def emergency_check(text):
    text = text.lower()
    for word in EMERGENCY_WORDS:
        if word in text:
            return True
    return False
```

```python
print(detect_emotion("I feel like ending my life"))
```

```python
chat_history = []

def chatbot_reply(user_input):
    chat_history.append(user_input)

    if emergency_check(user_input):
        return "üö® You are not alone.\nüìû India Helpline: 9152987821"

    sentiment, confidence = predict_sentiment(user_input)
    emotions = detect_emotion(user_input)

    response = f"""
Sentiment: {sentiment} ({confidence}%)
Top Emotions:
"""

    for emo, score in emotions:
        response += f"- {emo}: {score}%\n"

    return response
```

```python
while True:
    user = input("You: ")
    if user.lower() == "exit":
        break
    print("Bot:", chatbot_reply(user))
```

```python
import re

def clean_text(text):
    text = text.lower()
    text = re.sub(r"http\S+|www\S+", "", text)
    text = re.sub(r"[^a-z\s]", "", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text
```

```python
def predict_sentiment(text):
    if text is None or text.strip() == "":
        return "Neutral", 0.0   # default safe output

    text = clean_text(text)

    vec = vectorizer.transform([text])
    pred = sentiment_model.predict(vec)[0]
    prob = sentiment_model.predict_proba(vec).max() * 100

    label = "Positive" if pred == 1 else "Negative"

    # Neutral logic
    if prob < 60:
        label = "Neutral"

    return label, round(prob, 2)
# import re

# def clean_text(text):
#     text = text.lower()
#     text = re.sub(r"http\S+|www\S+", "", text)
#     text = re.sub(r"[^a-zA-Z\s]", "", text)
#     text = re.sub(r"\s+", " ", text).strip()
#     return text


```

```python
EMERGENCY_KEYWORDS = [
    "suicide", "kill myself", "end my life",
    "i want to die", "no reason to live",
    "self harm", "hurt myself"
]

def detect_emergency(text):
    if text is None or text.strip() == "":
        return False

    text = text.lower()
    for word in EMERGENCY_KEYWORDS:
        if word in text:
            return True
    return False
```

```python
def get_guidance(sentiment):
    guidance = {
        "Positive": "üòä Keep nurturing these positive moments. Stay connected with what makes you happy.",
        "Negative": "üíô I‚Äôm really glad you shared this. Try slow breathing or journaling your thoughts.",
        "Neutral": "üå± It‚Äôs okay to feel balanced. Small self-care habits can make a big difference.",
        "Emergency": "üö® You are not alone. Please contact a trusted person or helpline immediately."
    }
    return guidance.get(sentiment, "")
```
